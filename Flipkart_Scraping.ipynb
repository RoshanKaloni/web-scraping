```python
import csv

from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.flipkart.com/search?q=redmi+mobile&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_5_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_5_na_na_na&as-pos=2&as-type=RECENT&suggestionId=redmi+mobile%7CMobiles&requestId=0530d97f-244d-408e-9aa4-68d62d6a7daf&as-searchtext=redmi"
source = requests.get(url).text
```

```python
soup = BeautifulSoup(source, 'lxml')
```

```python
container = soup.find_all('div', class_="_2kHMtA")
```

```python
csv_file = open("flipkart_Scrap.csv", "w", encoding="utf-8")

csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Model_Name','Color', 'Memory & Storage Features', 'Display Features', 'Camera Features', 'Battery & Power Features', 'Os & Processor Features', 'Ratings', 'Price'])
```

```python
for title in container:
    product_name = title.div.img["alt"]
    title_name = product_name.rsplit(' (')
    model_Name = title_name.pop(0)              
    
    title_name = ''.join(title_name)
    title_name = title_name.rsplit(', ')
    color = title_name[0]                       
    
    
    extra_info = title.find_all('li', class_='rgWa7D')
    extra_info[:] = extra_info[:-1]
    
    Memory = extra_info[0].text                 
    
    Display = extra_info[1].text                
    
    Camera = extra_info[2].text                 
    
    Battery = extra_info[3].text                
    
    try:
        Processor = extra_info[4].text          
    except:
        Processor = None

    price = title.find("div", class_="_30jeq3 _1_WHN1").text[1:]   

    ratings = title.find("div", class_="_3LWZlK").text          


    csv_writer.writerow([model_Name, color, Memory, Display, Camera, Battery, Processor,  ratings, price])

csv_file.close()
```
